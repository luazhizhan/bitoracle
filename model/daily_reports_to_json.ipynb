```python
# load csv using pandas
import pandas as pd        
```

```python
# load monthly data
df = pd.read_csv('monthly_reports.csv')

# print MACD CROSS RF, MACD CROSS LSTM, MACD CROSS Bi-LSTM without the index to doll
print(df[['MACD CROSS RF', 'MACD CROSS LSTM', 'MACD CROSS Bi-LSTM']].round(2).to_string(index=False))

# save MACD CROSS LSTM to a csv file
df[['MACD CROSS Bi-LSTM']].round(2).to_csv('MACD_CROSS_Bi-LSTM.csv', index=False)
```
