```python
import pandas as pd
```

```python
df = pd.read_csv('data/data.csv')
columns_to_drop = ['BTC Active Addr Cnt', 'BTC Addr Cnt with ≥ .001% Supply', 'BTC Exchange Withdrawals', 
                   'BTC Exchange Deposits', 'BTC Miner Supply', 'BTC Miner Revenue', 'BTC Difficulty', 
                   'BTC Mean Block Size', 'BTC Total Fees']
df.drop(columns=columns_to_drop, inplace=True)
df.to_csv('data/data.csv', index=False)
```

```python
df = pd.read_csv('data/coinmetrics.csv')
df.rename(columns={'BTC / Active Addr Cnt': 'BTC Active Addr Cnt'}, inplace=True)
df.rename(columns={'BTC / Addr Cnt with ≥ .001% Supply': 'BTC Addr Cnt with ≥ .001% Supply'}, inplace=True)
df.rename(columns={'BTC / Exchange Withdrawals (USD)': 'BTC Exchange Withdrawals'}, inplace=True)
df.rename(columns={'BTC / Exchange Deposits (USD)': 'BTC Exchange Deposits'}, inplace=True)
df.rename(columns={'BTC / Miner Supply (USD)': 'BTC Miner Supply'}, inplace=True)
df.rename(columns={'BTC / Miner Revenue (USD)': 'BTC Miner Revenue'}, inplace=True)
df.rename(columns={'BTC / Difficulty': 'BTC Difficulty'}, inplace=True)
df.rename(columns={'BTC / Mean Block Size (in bytes)': 'BTC Mean Block Size'}, inplace=True)
df.rename(columns={'BTC / Total Fees (USD)': 'BTC Total Fees'}, inplace=True)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df
```

```python
df = df.sort_values(by='Date', ascending=True)
df
```

```python
# add one day to every row in Date column
df['Date'] = df['Date'] + pd.DateOffset(days=1)
```

```python
start_date = pd.to_datetime('2014-09-18')
end_date = pd.to_datetime('2025-04-20')

date_range = pd.date_range(start=start_date, end=end_date)
df = df.set_index('Date').reindex(date_range, method='nearest').reset_index()
df.rename(columns={'index': 'Date'}, inplace=True)
df
```

```python
df_data = pd.read_csv('data/data.csv')

# check if date column exists in data/data.csv
if 'Date' not in df_data.columns:
    # set Date column using df['Date']
    df_data['Date'] = df['Date']

if 'BTC Active Addr Cnt' not in df_data.columns:
    df_data['BTC Active Addr Cnt'] = df['BTC Active Addr Cnt']
if 'BTC Addr Cnt with ≥ .001% Supply' not in df_data.columns:
    df_data['BTC Addr Cnt with ≥ .001% Supply'] = df['BTC Addr Cnt with ≥ .001% Supply']
if 'BTC Exchange Withdrawals' not in df_data.columns:
    df_data['BTC Exchange Withdrawals'] = df['BTC Exchange Withdrawals']
if 'BTC Exchange Deposits' not in df_data.columns:
    df_data['BTC Exchange Deposits'] = df['BTC Exchange Deposits']
if 'BTC Miner Supply' not in df_data.columns:
    df_data['BTC Miner Supply'] = df['BTC Miner Supply']
if 'BTC Miner Revenue' not in df_data.columns:
    df_data['BTC Miner Revenue'] = df['BTC Miner Revenue']
if 'BTC Difficulty' not in df_data.columns:
    df_data['BTC Difficulty'] = df['BTC Difficulty']
if 'BTC Mean Block Size' not in df_data.columns:
    df_data['BTC Mean Block Size'] = df['BTC Mean Block Size']
if 'BTC Total Fees' not in df_data.columns:
    df_data['BTC Total Fees'] = df['BTC Total Fees']


df_data['Date'] = pd.to_datetime(df_data['Date'], format='%Y-%m-%d')

# check if there are missing newer dates in data/data.csv compared to date_range
if df_data['Date'].max() < df['Date'].max():
    # add missing dates to data/data.csv
    missing_dates = df[~df['Date'].isin(df_data['Date'])]
    missing_dates = missing_dates[['Date', 'BTC Active Addr Cnt', 'BTC Addr Cnt with ≥ .001% Supply', 
                                   'BTC Exchange Withdrawals', 'BTC Exchange Deposits', 'BTC Miner Supply', 
                                   'BTC Miner Revenue', 'BTC Difficulty', 'BTC Mean Block Size', 'BTC Total Fees']]
    df_data = pd.concat([df_data, missing_dates], ignore_index=True)

# write the updated data/data.csv
df_data.to_csv('data/data.csv', index=False)
```
